# SSL and TLS
* SSL - Secure Sockets Layer
* TLS - Transport Layer Security - most used security protocol on the internet (ex. HTTPS). It's created uniquely and on the spot between the client and server, keys are unique and secret. It uses AES to encrypt packets.
TLS supports RSA, AES, SHA encryption ecc.
Both are security transport protocols where SSL is the oldest type.

    * Session Key - usually are Simmetric Algorithms - present in TLS - is a key generated specifically for one session. Even if a session is compromised, all previews sessions are still secured by other session keys.
    * Cyphersuites - enumeration of prefered cryptography from a service provider, like Firefox (google it)

## SSL Stripping
Most common Man in the Middle attack, where an attacker poses as a Proxy and changes connections from HTTPS to HTTP.
* when interrogating a server, normally the first wave of traffic between the 2 sources is via HTTP, then a destionation server sends a 302 Redirect response to a HTTPS site that starts encryptions.
So the strip acts like a HTTP 302 response on behalf of the destination server and proxies all the connections. You can see this when you interrogate a known HTTPS web page and you get back a HTTP page - almost identical, the Green Lock is not Present! [kali has a stripper builtin]
* Techniques: 
    * ARP spoofing
    * Rougue AP, a wifi network that's set up to strip SSL: WiFi Pineapple

* Mitigation:
    * keep eye on the GreenLock
    * E2E encryption via VPN where all the traffic is passed to a secure gateway
    * using a L3 switch in an internal network

# HTTP & HTTPS
The main difference is when using HTTPS the browser will invoke SSL from the client to the server. The server authenticates the client via secure mutual authentication (Certificates) and estableshes a TLS connection (handshake).
In http obviously the S is not present so all of the traffic is being sent back&forward between the client and server via Plain-Text [that could be easily intercepted]
* Issuing a Broken Certificate (CA) into a victim browser may lead to information leaks, thus hacked web browsers -> knowing the Private Key of a Certificate will eventually decrypt the traffic.
* SSLsniff, if you're in the middle, you can create a bogus CA so the browser can trust and push private traffic through it. The browser can be very vulnerable if trusting bad CAs.
CAs change all the time.
* Key Pinning - the host gets pinned with a specific public key or CA so it cannot be changed by MITM attacks. It can be used with VPN and/or TLS.
