# Malware
is all the software that have malicious intent:
* Macro Virus = imbedded VBScrit (Microsoft Visual Basic Script) in MS Office Software
* Stealth Virus = intercepts and tricks antivirus sw and keep exploiting
* Polymorphic Virus = makes copies of itself and changes for based on the OS is running on so it makes it harder to detect and mitigate
* Self-garbling Virus = changes it's code so antivirus sw cannot detect it based on previous code signatures
* Bots & Zombies = compromised machines that operate to do specific tasks like crypto mining
* Worms = virus kits that spread from a machine to the other
* OS Rootkit = imbedded in the kernerl of the OS, worst kind of virus because when ran, it gives root access to that specific machine and therefore has access to all of the files and documents
* Firmware Rootkit = worst of all, once placed into memory of a machine, it doesn't go away neighter by reinstalling the OS -> NSA grade stuff
* Keylogger = they log key strokes
* Trojan Horses = software that act like one software but does malicious actions in background
* Remote Access Tool (RAT) = like remote administration tools, they give remote access 
* Ransomware = malware taking control over machine, encrypts files and asking for ransom for file decryption
* Malvetisement = online advertisment platforms host ads that have scripts that point to other scripts that eventually turn into malicious code -> growing attack trend
* Drive-by Attacks = hackers put malicious script into vulnerable websites to spread malware on site visitors
* Spyware = software that gather information without damaging a specific user, just breaking his anonymity
* Adware = undesired software that forces ads over you and actively works itself around by trying not to be removed from the machine [browser-hijacking]
* Scareware = fake warning pop-ups that show on machine claming to fix a problem that doesn't exit
* Potentially Unwanted Programs (PUPs) = the type of programs that you maybe consider unwanted, but your antivirus sw doesn't understand if you want to install that program or not
* Phishing = a form of social engineering; tricks a victim to execute(clicks) malware on machine, can be an attempt to gain access to the machine -> most successful, cheap to run, high yeald, common
    * Spear Phishing = when you are the only target of the Phishing attack [so phishing attacks are done in mass]

# CPU hijackers
* On the rise of popularity is the hijacking of machines to generate cryptocurrency via CPU mining.
* Almost all malware can be used to inject a miner on a victim machine. 
* Another way to accomplish this is via JavaScript off Browsers or Websites. Some websites do this but with user notification.
* Android malware is a new trend for CPU hijacking, Apple is safer

## CPU hijacking mitigation
Keep a close eye on the processes on Task/Activity Manager. Install NoCoin or MinerBlock on Browsers.
